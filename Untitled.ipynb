```python
def plot_it(preds2, plot_col=1, plot_step=14):
    if type(preds2) is not np.ndarray:
        # pick a column to plot.
        predvals = preds2[0]
        pred_logitsigs = preds2[1] * .49 + .5
        predsigs = pred_logitsigs / (1 - pred_logitsigs)
    else:
        predvals = preds2
        predsigs = predvals * 0.

    print("[plotting...]")
    print("x: %s, preds: %s" % (x.shape, predvals.shape))


    my_order = np.argsort(x[:,plot_step,plot_col])
    difs = x[my_order,plot_step,plot_col] - predvals[my_order,plot_step,plot_col]
    zs = np.abs(difs / predsigs[my_order,plot_step,plot_col])

    plt.scatter(y=predvals[my_order,plot_step,plot_col] - predsigs[my_order,plot_step,plot_col], s=1, x=range(len(my_order)), label='pred_low', alpha=.1)
    plt.scatter(y=predvals[my_order,plot_step,plot_col] + predsigs[my_order,plot_step,plot_col], s=1, x=range(len(my_order)), label='pred_hi', alpha=.1)
    plt.scatter(y=predvals[my_order,plot_step,plot_col], x=range(len(my_order)), s=1, label='predict', alpha=.02)
    plt.scatter(y=x[my_order,plot_step,plot_col], x=range(len(my_order)), s=1, label='data', alpha=.1)
    plt.scatter(y=np.clip(abs(zs),0,1.5), x=range(len(my_order)), s=1, label='z-scores', alpha=.05)


    plt.legend()
    plt.show()
    return(plt)

def pred_and_weights(vae,enc,gen,x,batch_size):
    preds = vae.predict(x, batch_size=batch_size)
    vweights = [layer.get_weights() for layer in vae.layers]
    eweights = [layer.get_weights() for layer in enc.layers]
    gweights = [layer.get_weights() for layer in gen.layers]
    return (preds, (vweights, eweights, gweights))

def set_weights(vae,enc,gen,weights):
    v,e,g = weights
    for lweights, layer in zip(g,gen.layers):
        layer.set_weights(lweights)
    for lweights, layer in zip(e,enc.layers):
        layer.set_weights(lweights)
    for lweights, layer in zip(v,vae.layers):
        layer.set_weights(lweights)
    
DO_SET_WEIGHTS = False
    
```

```python

from keras import backend as K
from keras.callbacks import Callback

EON_EPOCHS = 30
KL_START_EPOCH = 5
KL_ANNEAL_TIME = 9 #so 10 epochs of KL weight=min, 18 to anneal, ~30 epochs of weight=1, then repeat.
MIN_KL_WEIGHT = 0.1


class AnnealingCallback(Callback):
    def __init__(self, weight):
        self.weight = weight
    def on_epoch_end (self, epoch, logs={}):
        effective_epoch = epoch % EON_EPOCHS
        if effective_epoch == 0:
            K.set_value(self.weight, MIN_KL_WEIGHT)
        if (effective_epoch) >= KL_START_EPOCH :
            new_weight = min(K.get_value(self.weight) + ((1. - MIN_KL_WEIGHT)/ KL_ANNEAL_TIME), 1.)
            K.set_value(self.weight, new_weight)
        print (" KLfac = " + str(K.get_value(self.weight)))
        
cb2 = AnnealingCallback(cb.weight)
```
