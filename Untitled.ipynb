```python
pages = list()
page = list()
cur_id = ""
cur_pnum = -1
for row in raw.itertuples():
    if row.file_id != cur_id or row.page != cur_pnum:
        #print(f"{row.file_id} {row.page}")
        if page:
            pages.append(page)
            page = list()
        cur_id = row.file_id
        cur_pnum = row.page
    page.append([np.log(row.len) / 4,row.left,row.offset,row.height,row.width,row.confidence / 100,row.near_bottom])
    
len(pages)
        
```

```python
def plot_it(preds2, plot_col=1, plot_step=14):
    if type(preds2) is not np.ndarray:
        # pick a column to plot.
        predvals = preds2[0]
        pred_logitsigs = preds2[1] * .49 + .5
        predsigs = pred_logitsigs / (1 - pred_logitsigs)
    else:
        predvals = preds2
        predsigs = predvals * 0.

    print("[plotting...]")
    print("x: %s, preds: %s" % (x.shape, predvals.shape))


    my_order = np.argsort(x[:,plot_step,plot_col])

    plt.scatter(y=predvals[my_order,plot_step,plot_col] - predsigs[:,plot_step,plot_col], s=1, x=range(len(my_order)), label='pred_low', alpha=.1)
    plt.scatter(y=predvals[my_order,plot_step,plot_col] + predsigs[:,plot_step,plot_col], s=1, x=range(len(my_order)), label='pred_hi', alpha=.1)
    plt.scatter(y=predvals[my_order,plot_step,plot_col], x=range(len(my_order)), s=1, label='predict', alpha=.02)
    plt.scatter(y=x[my_order,plot_step,plot_col], x=range(len(my_order)), s=1, label='data', alpha=.1)


    plt.legend()
    plt.show()
    return(plt)

plot_it(preds2)

#107
```
